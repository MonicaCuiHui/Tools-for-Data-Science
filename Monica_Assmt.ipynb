```python
import matplotlib.pyplot as plt
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])

end_date = df['DATE'].max()
start_date = end_date - pd.DateOffset(days=365)

df_recent = df[(df['DATE'] >= start_date) & (df['DATE'] <= end_date)]

df_weekly = df_recent.groupby([pd.Grouper(key='DATE', freq='W'), 'age_category']).first().reset_index()

print(df_weekly.head())

```

```python
# Plot the data
plt.figure(figsize=(12, 6))

# Loop through each age category and plot its weekly trend
for age_category in df_weekly["age_category"].unique():
    df_age = df_weekly[df_weekly["age_category"] == age_category]
    plt.plot(df_age["DATE"], df_age["percent_positive_7d_avg"], 
             marker='o', linestyle='-', label=f"Age Category: {age_category}")
    
# Labels and formatting
plt.xlabel("Date")
plt.ylabel("Percent Positivity (%)")
plt.title("Weekly Percent Positivity by Age Category (Last 1 Year)")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

# Display the legend with age category labels
plt.legend(title="Age Category")
plt.tight_layout()
plt.show()
```
